```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
# Upload: StudentsPerformance.csv and gym_membership.csv

# Load data
#students = pd.read_csv('StudentsPerformance.csv')
gym = pd.read_csv('gym_membership.csv')
```

```python
#1. Bar chart of gym membership types
bar = gym['abonoment_type'].value_counts().plot(kind='bar')
plt.title('Bar chart of gym membership types')
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Histogram of time in gym
plt.hist(gym['avg_time_in_gym'], bins= 30, color='orange')
plt.title('Histogram of time in gym')
plt.xlabel('Time in gym')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title('Scatter plot: Age vs visits per week')
plt.xlabel('Age')
plt.ylabel('Visits per week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
sns.boxplot(data=gym, x='gender', y='Age')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Age by Gender')
plt.show()
```

```python
# 2. Count plot: Days per week
sns.countplot(data=gym, x='days_per_week', color='orange', order=gym['days_per_week'].value_counts().index)
plt.title('gym/days_per_week')
plt.show()
```

```python
# 3. Bar plot: Average visits by membership type
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average visits by membership type')
plt.show()
```
